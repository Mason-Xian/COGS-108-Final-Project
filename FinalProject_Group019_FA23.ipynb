# Abstract

Our project compares patient-rated effectiveness and satisfaction between telehealth and in-person care in the post-pandemic United States. Our descriptive and visual data exploration analysis of patient surveys reveals a higher preference for telehealth, attributed to its convenience and broader access, particularly for uninsured individuals. Despite the general favorability, a subset of patients expresses a preference for traditional in-person encounters, highlighting the need for personalized healthcare options. Further research is recommended to explore nuanced patient preferences and optimize telehealth integration within the healthcare system.

# Research Question

In post-pandemic United States, how does the effectiveness of telemedicine compare to traditional in-person sessions for those patients who have telemedicine available to them?
- The operational definition of “the effectiveness of telemedicine” is patient satisfaction scores on surveys

## Background and Prior Work

The landscape of health care has undergone significant transformations, particularly with the increasing adoption of telehealth, intensified by the COVID-19 pandemic. Nowadays, patients can receive telemedicine care for both mental and physical ailments. This shift has prompted several studies to scrutinize the comparative effectiveness of virtual care sessions against traditional in-person interventions.

"In‐person versus virtual therapy in outpatient eating‐disorder treatment: A COVID‐19 inspired study" investigated the outcomes of individuals undergoing in-person and virtual therapy during the COVID-19 outbreak. Results indicated similar improvements in eating disorder symptoms, weight gain, and satisfaction with services in both groups<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5).

Lee et al. explored the patient experience of telehealth services in a mental health setting during the COVID-19 pandemic, employing a cross-sectional survey conducted in an outpatient psychiatric clinic<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). Their findings contribute valuable insights into how patients perceive telehealth services in mental health care.

Comparing telehealth and face-to-face care for less common mental health conditions, Greenwood et al. aimed to assess differences in effectiveness between the two treatment modalities<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). 

Polinski et al. delved into patient satisfaction and preferences with telehealth visits through a cross-sectional survey involving 1,734 patients<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). The results underscore high patient satisfaction, with 94-99% reporting being "very satisfied" with various aspects of telehealth. Predictors of satisfaction included gender, overall understanding, quality of care, and convenience of telehealth.

Kaur et al. explored patient satisfaction rates during the COVID-19 era, emphasizing the importance of communication methods in telemedicine<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). The study found that satisfaction rates were higher when video calls were used for teleconsultations, signaling the significance of the mode of interaction in influencing patient satisfaction.

Collectively, these studies suggest that telemedicine has the potential to be as effective as traditional in-person sessions for certain medical care. The high satisfaction rates reported by patients further highlight the convenience and perceived quality of care associated with telehealth, contributing to improved accessibility to mental health services and physical consultations alike.

In our study, we want to explore the possible difference of effectiveness and satisfaction that comes with the adoption of telehealth. We will use data analysis tools to answer our question: how does the effectiveness of telemedicine compare to traditional in-person sessions for those patients who have telemedicine available to them the post-COVID 19 America?

### SU63a Opiod Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
filepath = "csv-corrected/SU63a-opioids.csv"
cont_opioids_filepath = "csv-corrected/SU63a-cont-opioids.csv"

opioids_data = pd.read_csv(filepath, skiprows=2)
cont_opioids_data = pd.read_csv(cont_opioids_filepath, skiprows=1)

# Select the relevant columns for the plot using index positions
states_opioids = opioids_data.iloc[:, 0]
telehealth_values_opioids = opioids_data.iloc[:, 13]

states_cont_opioids = cont_opioids_data.iloc[:, 0]
telehealth_values_cont_opioids = cont_opioids_data.iloc[:, 13]

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(states_opioids, telehealth_values_opioids, color='green')
plt.bar(states_cont_opioids, telehealth_values_cont_opioids, color='green')

plt.title('Usage of Telehealth in Substance Use Treatment Facilities by State')
plt.xlabel('State or jurisdiction')
plt.ylabel('Number of Facilities Using Telehealth')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```

As it turns out, state with more facilities that provide telehealth also have more cases of relapse prevention.
